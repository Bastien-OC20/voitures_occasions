```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd

# Initialiser le driver
options = webdriver.ChromeOptions()
options.add_argument("--disable-blink-features=AutomationControlled")  # Pour essayer de contourner les dtections automatiques
options.add_argument("--headless")  # Excute Chrome en mode headless pour plus de discrtion
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
driver.get("https://www.aramisauto.com/achat/occasion")

# Attendre le chargement de la page
time.sleep(3)

# Initialiser la pagination
current_page = 1
voitures = []

while True:
    try:
        print(f"Scraping page {current_page}")
        
        # Trouver tous les conteneurs des annonces
        annonces = driver.find_elements(By.CLASS_NAME, "item")

        # Extraire les informations de chaque annonce
        for annonce in annonces:
            try:
                # Extraire la marque et le modle
                # marque_modele = annonce.find_element(By.CLASS_NAME, 'product-card-vehicle-information__title').text

                # Extraire le prix
                prix = annonce.find_element(By.CLASS_NAME, 'heading-l').text

                # Extraire les autres informations (anne, kilomtrage, type de carburant)
                details_brut = annonce.find_element(By.CLASS_NAME, 'product-card-vehicle-information__bottom').text

                # Diviser les dtails en fonction des sparateurs visuels (ici "")
                details = details_brut.split("\u2022")

                # Nettoyer les espaces superflus
                details = [detail.strip() for detail in details]

                # Extraire chaque information (en vrifiant la longueur de la liste)
                annee = details[0] if len(details) > 0 else "Non spcifi"
                kilometrage = details[1] if len(details) > 1 else "Non spcifi"
                etat = details[2] if len(details) > 2 else "Non spcifi"

                voitures.append({
                    'Marque/Modle': marque_modele,
                    'Anne': annee,
                    'Kilomtrage': kilometrage,
                    'Etat': etat,
                    'Prix': prix
                })
            except Exception as e:
                print(f"Erreur lors de l'extraction des informations d'une annonce: {e}")

        # Passer  la page suivante en cliquant sur le bouton suivant
        try:
            next_button = driver.find_element(By.CLASS_NAME, "element")
            print(next_button)
            next_button.click()
            time.sleep(3)
            current_page += 1
        except Exception as e:
            print(f"Erreur lors de la navigation vers la page suivante ou fin des pages: {e}")
            break

    except Exception as e:
        print(f"Erreur lors de la rcupration des annonces: {e}")
        break

# Crer un DataFrame et sauvegarder dans un fichier CSV
df = pd.DataFrame(voitures)
df.to_csv('ouiiiiiiiiii.csv', index=False)

# Fermer le driver
driver.quit()
```
