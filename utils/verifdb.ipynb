```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect("../data/db/voitures_aramisauto.db")

# Create a cursor object
cursor = conn.cursor()

# Vérifier les marques
cursor.execute("SELECT * FROM marque WHERE ID_Marque = 65")
marques = cursor.fetchall()
print("Marques:", marques)

# Vérifier les véhicules associés à la marque
cursor.execute("SELECT * FROM vehicule WHERE Marque_ID = 114")
vehicules = cursor.fetchall()
print("Véhicules:", vehicules)

# Vérifier les modèles associés aux véhicules
cursor.execute(
    """
SELECT DISTINCT m.Nom 
FROM vehicule v
JOIN modele m ON v.Modele_ID = m.ID_Modele
WHERE v.Marque_ID = 114
"""
)
modeles = cursor.fetchall()
print("Modèles:", modeles)

# Vérifier les marques et modèles associés aux véhicules
cursor.execute(
    """
SELECT DISTINCT m.Nom, ma.Nom 
FROM vehicule v
JOIN modele m ON v.Modele_ID = m.ID_Modele
JOIN marque ma ON v.Marque_ID = ma.ID_Marque
WHERE v.Marque_ID = 114
"""
)
marques_modeles = cursor.fetchall()
print("Marques et Modèles:", marques_modeles)

# Close the connection
conn.close()
```

```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect("../data/db/voitures_aramisauto.db")

# Create a cursor object
cursor = conn.cursor()

# Lire toutes les lignes de la table vehicule
cursor.execute("SELECT * FROM vehicule")
vehicules = cursor.fetchall()

# Lire toutes les lignes de la table modele
cursor.execute("SELECT * FROM modele")
modeles = cursor.fetchall()

# Lire toutes les lignes de la table marque
cursor.execute("SELECT * FROM marque")
marques = cursor.fetchall()

# Créer des dictionnaires pour les modèles et les marques
modele_dict = {modele[0]: modele[1] for modele in modeles}
marque_dict = {marque[0]: marque[1] for marque in marques}

# Associer les modèles aux marques en utilisant les informations de la table vehicule
resultats = []
for vehicule in vehicules:
    modele_id = vehicule[1]
    marque_id = vehicule[5]
    modele_nom = modele_dict.get(modele_id, "Inconnu")
    marque_nom = marque_dict.get(marque_id, "Inconnu")
    resultats.append((modele_nom, marque_nom))

# Afficher les résultats
for resultat in resultats:
    print("Modèle:", resultat[0], "| Marque:", resultat[1])

# Close the connection
conn.close()
```
